```python
Lat = pd.cut(Meteo["LAT"], [Meteo["LAT"].min()-1, 0, Meteo["LAT"].max()+1], right=False)
```

```python
Meteo.groupby(Lat)["WBAN"].count()
```

```python
Meteo[ Meteo["LAT"].isna() ]["WBAN"].count()
```

```python
Meteo["BEGIN"] = pd.to_datetime(Meteo["BEGIN"], format='%Y%m%d')
Meteo["END"] = pd.to_datetime(Meteo["END"], format='%Y%m%d')
```

```python
Meteo.head(5)
```

```python
Meteo["ACT"] = Meteo["END"]-Meteo["BEGIN"]
```
