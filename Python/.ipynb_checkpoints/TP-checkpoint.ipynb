```python
import numpy as np
import pandas as pd
import numpy.random as npr 
```

```python
n=10
ind = range(1, n+1) #pour cr√©er une liste 
noms = [f"enregistrement {i}" for i in ind]

val1 = [ npr.random() for i in ind]
val2= [ npr.random() for i in ind]
val3= [ npr.random() for i in ind]

Snoms = pd.Series(noms,index=ind)
Sval1 = pd.Series(val1, index = ind)
Sval2 = pd.Series(val2,index=ind)
Sval3 = pd.Series(val3, index=ind)

dic = {
    'Nom': Snoms,
    'val1':Sval1,
    'val2': Sval2,
    'val3' :Sval3
}
```

```python
df = pd.DataFrame(dic)
```

```python
df['Moy'] = (df['val1'] + df["val2"] + df['val3'])/3
df

df.drop('val2',inplace=True, axis=1)
```

```python
df
```

```python
#les filtres

df2 = df.copy()
df2 = df[df['Moy']>=0.5] #a revoir les masques 
df2
```
